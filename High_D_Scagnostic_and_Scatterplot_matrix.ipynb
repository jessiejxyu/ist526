# Interpret Scagnostic
[library](https://github.com/haishiro/pyscagnostics)

## Install library

```
# installation - run it once
!pip install pyscagnostics
```

## loading + clearning data

```
from pyscagnostics import scagnostics

## make sure all columns are numeric
display(df.head())
```

## Get scores for the data frame

```
# get sgagnostic scores from the library
results = scagnostics(df)
```

## Convert scores to Scagnostic-format

```
# post-process results to make scagnostic

# create 9-attribute list
score_names = ['Outlying', 'Skewed', 'Clumpy', 'Sparse', 'Striated', 'Convex', 'Skinny', 'Stringy', 'Monotonic']

# add a new data source
data = {'score_names':score_names}

scores = None

# iterate over the reasults
for i, j, result in results:
  (measures, bins) = result  
  scores = measures
  
  ## debug
  print(i, j, measures)
  
  col_name = "{}_{}".format(i, j)  
  
  # copy scores
  row = []
  row.append(scores['Outlying'])
  row.append(scores['Skewed'])
  row.append(scores['Clumpy'])
  row.append(scores['Sparse'])
  row.append(scores['Striated'])
  row.append(scores['Convex'])
  row.append(scores['Skinny'])
  row.append(scores['Stringy'])
  row.append(scores['Monotonic'])

  # add current 'col_name' as key and 'row' as value to 'data'
  data[col_name] = row;
# end for
```

## Create a dataframe for scagnostic visualization

```
# convert 'data' to data-frame named 'df_scagnostic'
df_scagnostic = pd.DataFrame.from_dict(data);

# make the first column 'score_names' to index
df_scagnostic.set_index('score_names', inplace=True)

# check out the dataframe
display(df_scagnostic.head(9))
```

## Finally, visualize scagnostic

```
# now that our dataframe is in perfect condition, visualiza it with plotly
import plotly.express as px

fig = px.imshow(df_scagnostic, text_auto=True, aspect="auto")
fig.update_xaxes(side="top")
fig.show()
```
